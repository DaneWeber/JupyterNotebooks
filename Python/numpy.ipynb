```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
```

```python
class DividedTriangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.leftx = 0
        self.rightx = self.leftx + width
        self.centerx = self.leftx + (width / 2)
        self.bottomy = 0
        self.topy = self.bottomy + self.height
        self.waterliney = self.bottomy + 1

        slope = (width / 2) / height
        self.leftwaterlinex = self.leftx + slope
        self.rightwaterlinex = self.rightx - slope

        totalarea = width * height / 2
        self.toparea = (width - (2 * slope)) * (height - 1) / 2
        self.basearea = totalarea - self.toparea

    def full(self):
        return Polygon(np.array([[self.leftx, self.bottomy], [self.centerx, self.topy], [self.rightx, self.bottomy]]), closed=False, fc='gray')

    def base(self):
        return Polygon(np.array([[self.leftx, self.bottomy], [self.leftwaterlinex, self.waterliney], [self.rightwaterlinex, self.waterliney], [self.rightx, self.bottomy]]), closed=False, fc='green')

    def label(self, plotspace):
        bottompad = 0.2
        plotspace.annotate(self.basearea, [self.centerx, self.bottomy + bottompad], ha='center', color='white')
        plotspace.annotate(self.toparea, [self.centerx, self.waterliney + bottompad], ha='center', color='white')
```

```python
dt = DividedTriangle(2, 2)
ax = plt.gca()
ax.add_patch(dt.full())
ax.add_patch(dt.base())
dt.label(ax)
ax.set_xlim(-1,7)
ax.set_ylim(-1,7)
plt.show()
```
